# Combining Matrixes and Weighting Them

```python
tfidf_vectorizers = [tfidf_des, tfidf_char, tfidf_concept, 
                    tfidf_location, tfidf_object, tfidf_person,
                    tfidf_arc, tfidf_team, tfidf_vol]
```

```python
all_matrices = [tfidf_des_matrix, tfidf_char_matrix, tfidf_concept_matrix, 
               tfidf_location_matrix, tfidf_object_matrix, tfidf_person_matrix,
               tfidf_arc_matrix, tfidf_team_matrix, tfidf_vol_matrix]
```

```python
weights = [1, 1, 1,
           1, 1, 1,
           1, 1, 1]
```

```python
# Scale each TF-IDF matrix by its corresponding weight
weighted_tfidf_matrices = [matrix * weight for matrix, weight in zip(all_matrices, weights)]
```

```python
combined_matrix = hstack(weighted_tfidf_matrices)
```

# COSINE_SIM & RECOMMENDATION FUNCTION

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim_combined = linear_kernel(combined_matrix, combined_matrix)
```

```python
def get_recommendations(df, title, cosine_sim=cosine_sim):
    # Get the index of the issue that matches the title
    idx = df[df['name'] == title].index[0]
    # Get the pairwise similarity scores of all issues with that issue
    sim_scores = list(enumerate(cosine_sim[idx]))
    # Sort the issues based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get the scores of the 10 most similar issues
    sim_scores = sim_scores[1:20]
    # Get the issue indices
    issue_indices = [i[0] for i in sim_scores]
    # Return the top 10 most similar issues
    return df[['name', 'issue_number', 'description', 'character_credits','volume']].iloc[issue_indices]
```

```python
rec = get_recommendations(df, "Welcome to Gotham Academy", cosine_sim)
rec_combined = get_recommendations(df, "Welcome to Gotham Academy", cosine_sim_combined)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

# Sample DataFrame with 9 text columns
data = {
    'desc1': ["This is a sample description.", "Another sample description.", "Yet another description."],
    'desc2': ["Some more text.", "Different text here.", "More and more text."],
    'desc3': ["Text data.", "Data text.", "Sample data."],
    'desc4': ["Additional text here.", "Another line of text.", "More sample text."],
    'desc5': ["Yet more descriptions.", "Even more descriptions.", "Descriptions galore."],
    'desc6': ["Some keywords.", "More keywords.", "Yet more keywords."],
    'desc7': ["Extra text.", "Additional content.", "Supplementary text."],
    'desc8': ["Content goes here.", "More content.", "Extra content."],
    'desc9': ["Final piece of text.", "Conclusion text.", "Ending text."]
}

df = pd.DataFrame(data)

# Initialize TfidfVectorizer for each column
tfidf_vectorizers = [TfidfVectorizer() for _ in range(9)]

# Fit and transform each column
tfidf_matrices = [vectorizer.fit_transform(df[f'desc{i+1}']) for i, vectorizer in enumerate(tfidf_vectorizers)]

# Define weights for each column
weights = [1, 2, 1, 1, 2, 1, 1, 1, 2]  # Adjust these weights as needed

# Scale each TF-IDF matrix by its corresponding weight
weighted_tfidf_matrices = [matrix * weight for matrix, weight in zip(tfidf_matrices, weights)]

# Concatenate the resulting matrices
combined_matrix = hstack(weighted_tfidf_matrices)

# Get feature names from each vectorizer
feature_names = []
for i, vectorizer in enumerate(tfidf_vectorizers):
    feature_names.extend([f'desc{i+1}_' + feature for feature in vectorizer.get_feature_names_out()])

print(combined_matrix)
print(feature_names)
```
